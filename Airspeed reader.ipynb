```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("First_Flight.txt")
df2 = pd.read_csv("2nd_Flight_Crash.TXT")
```

```python
def moving_average(a, n=50) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
time1 = df1["time"]
yaw1 = df1["yaw"]
pitch1 = df1["pitch"]
roll1 = df1["roll"]
volts1 = df1["voltage"]
temp1 = df1["temperature"]

plt.plot(time1, yaw1, label = "Yaw")
plt.plot(time1, roll1, label = "Roll")
plt.plot(time1, pitch1, label = "Pitch")
plt.legend()
plt.title("Orientation for Flight 1")
plt.show()
plt.plot(time1, volts1)
plt.title("Voltage over time - Flight 1")
plt.show()
plt.plot(time1, temp1)
plt.title("Temperature - Flight 1")
plt.show()
```

```python
time2 = df2["time"]
yaw2 = df2["yaw"]
pitch2 = df2["pitch"]
roll2 = df2["roll"]
volts2 = df2["voltage"]
temp2 = df2["temperature"]

plt.plot(time2, yaw2, label = "Yaw")
plt.plot(time2, roll2, label = "Roll")
plt.plot(time2, pitch2, label = "Pitch")
plt.legend()
plt.title("Orientation Flight 2 - crash")
plt.show()
plt.plot(time2, volts2)
plt.title("Voltage Flight 2 - crash")
plt.show()
plt.plot(time2, temp2)
plt.title("Temperature Flight 2 - crash")
plt.show()
```
