```python
def moving_average(a, n=50) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
data1 = df1["data"].to_numpy()
time1 = df1["time"].to_numpy()
#plt.scatter(moving_average(time1[1:10000]), moving_average(data1[1:10000]))
plt.scatter(time1, data1)
plt.show()
```
